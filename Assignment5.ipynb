
# Train a Random Forest model



```python
# Let's first load the uploaded CSV file to understand its structure before we proceed with splitting the dataset and training the models.
import pandas as pd

# Load the dataset
file_path = '/content/sales_data.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# Preprocessing for numerical features
numerical_transformer = SimpleImputer(strategy='median')

# Preprocessing for categorical features
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Columns that will be considered in the model
numerical_cols = data.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = data.select_dtypes(include=['object']).columns.drop('flag')  # Exclude target
```

```python
# Bundle preprocessing for numerical and categorical data
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])

# Define the model
model = RandomForestClassifier(n_estimators=100, random_state=0)
```

```python
# Bundle preprocessing and modeling code in a pipeline
clf = Pipeline(steps=[('preprocessor', preprocessor),
                      ('model', model)])
```

```python
# Separate target from predictors
y = data.flag
X = data.drop('flag', axis=1)

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Preprocessing and training the model
clf.fit(X_train, y_train)

# Output the shape of the train and test sets as a simple check
X_train.shape, X_test.shape
```

```python
from sklearn.metrics import accuracy_score

# Making predictions on the test set
y_pred = clf.predict(X_test)

# Evaluating the model
accuracy = accuracy_score(y_test, y_pred)

print(f"Model accuracy on the test set: {accuracy:.2f}")
```

```python
from sklearn.model_selection import GridSearchCV

# Parameters to tune
param_grid = {
    'model__n_estimators': [10, 20, 30],  # Number of trees
    'model__max_depth': [None, 5, 10, 15]  # Maximum depth of trees
}

# Create a GridSearchCV object
grid_search = GridSearchCV(estimator=clf, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)

# Fit the model with the grid search
grid_search.fit(X_train, y_train)

# Print the best parameters and the best score
print("Best parameters found: ", grid_search.best_params_)
print("Best score found: ", grid_search.best_score_)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Original model predictions
original_y_pred = clf.predict(X_test)

# Best model from grid search predictions
tuned_y_pred = grid_search.predict(X_test)

# Function to calculate metrics
def evaluate_model(y_true, y_pred, model_name="Model"):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, pos_label='Y')
    recall = recall_score(y_true, y_pred, pos_label='Y')
    f1 = f1_score(y_true, y_pred, pos_label='Y')

    print(f"{model_name}:")
    print(f"  Accuracy: {accuracy:.4f}")
    print(f"  Precision: {precision:.4f}")
    print(f"  Recall: {recall:.4f}")
    print(f"  F1 Score: {f1:.4f}")
    print()

# Evaluate original model
evaluate_model(y_test, original_y_pred, "Original Model")

# Evaluate tuned model
evaluate_model(y_test, tuned_y_pred, "Tuned Model")
```

# Train a Support Vector Machine

```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC

# Load your dataset. Ensure you replace the path with where your dataset is located in Google Colab
data = pd.read_csv('/content/sales_data.csv')  # Adjust this path to your dataset location
```

```python
# Setup preprocessing steps
# For numerical features, fill missing values with the median of each column
numerical_transformer = SimpleImputer(strategy='median')

# For categorical features, fill missing values with the most frequent value and apply one-hot encoding
categorical_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
# Identify the numerical and categorical columns, excluding the target variable 'flag'
numerical_cols = data.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = data.select_dtypes(include=['object']).columns.drop('flag')
```

```python
# Bundle the preprocessing for both types of features
preprocessor = ColumnTransformer([
    ('num', numerical_transformer, numerical_cols),
    ('cat', categorical_transformer, categorical_cols)
])
```

```python
# Define the model to be used, in this case, Support Vector Machine (SVC)
model = SVC(random_state=0)

# Create a pipeline that encapsulates preprocessing and the model
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', model)
])

# Prepare data for training and testing
y = data['flag']  # Target variable
X = data.drop('flag', axis=1)  # Features

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train the model on the training data
pipeline.fit(X_train, y_train)

# The model is now trained and can be used to make predictions
# Example:
# y_pred = pipeline.predict(X_test)
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid to search over
param_grid = {
    'model__C': [0.1, 1, 10],  # Regularization parameter
    'model__gamma': ['scale', 'auto'],  # Kernel coefficient
}

# Create a GridSearchCV object with the pipeline and the parameter grid
# Use cv=5 for 5-fold cross-validation and n_jobs=-1 to use all available cores
grid_search = GridSearchCV(pipeline, param_grid, cv=5, n_jobs=-1, verbose=2)

# Fit the model with grid search
grid_search.fit(X_train, y_train)

# Print the best parameters and the corresponding score
print("Best parameters found:", grid_search.best_params_)
print("Best cross-validation score:", grid_search.best_score_)

# The grid_search object is now a trained model with the best found parameters
# You can use it to make predictions or further evaluate its performance
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Make predictions with the original model
y_pred_original = pipeline.predict(X_test)

# Make predictions with the tuned model
y_pred_tuned = grid_search.predict(X_test)

# Define a function to calculate and print the metrics
def print_performance_metrics(y_true, y_pred, model_name):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, pos_label='Y')
    recall = recall_score(y_true, y_pred, pos_label='Y')
    f1 = f1_score(y_true, y_pred, pos_label='Y')

    print(f"{model_name} performance:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}\n")

# Evaluate the original model
print_performance_metrics(y_test, y_pred_original, "Original SVM Model")

# Evaluate the tuned model
print_performance_metrics(y_test, y_pred_tuned, "Tuned SVM Model")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Assuming y_test, y_pred_original, and y_pred_tuned are defined from previous model predictions

# Calculate performance metrics for both models
def calculate_metrics(y_true, y_pred):
    metrics = {}
    metrics['Accuracy'] = accuracy_score(y_true, y_pred)
    metrics['Precision'] = precision_score(y_true, y_pred, pos_label='Y')
    metrics['Recall'] = recall_score(y_true, y_pred, pos_label='Y')
    metrics['F1 Score'] = f1_score(y_true, y_pred, pos_label='Y')
    return metrics

original_metrics = calculate_metrics(y_test, y_pred_original)
tuned_metrics = calculate_metrics(y_test, y_pred_tuned)

# Plotting metrics for original and tuned models
metrics_names = list(original_metrics.keys())
original_values = list(original_metrics.values())
tuned_values = list(tuned_metrics.values())

fig, axs = plt.subplots(1, 5, figsize=(20, 5))  # Creating 5 subplots

# Plotting the first four metrics
for i, metric in enumerate(metrics_names):
    axs[i].bar(['Original', 'Tuned'], [original_values[i], tuned_values[i]], color=['skyblue', 'orange'])
    axs[i].set_title(metric)

# Confusion Matrix for the Tuned Model
cm = confusion_matrix(y_test, y_pred_tuned, labels=['Y', 'N'])
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axs[4])
axs[4].set_title('Confusion Matrix (Tuned Model)')
axs[4].set_xlabel('Predicted Labels')
axs[4].set_ylabel('True Labels')
axs[4].set_xticklabels(['Y', 'N'])
axs[4].set_yticklabels(['Y', 'N'], rotation=0)

plt.tight_layout()
plt.show()
```
